```python
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
url = "https://raw.githubusercontent.com/murpi/wilddata/master/quests/churn_telecom.csv"
data = pd.read_csv(url)
df = pd.DataFrame(data)
df.head()
```

```python
df.shape
```

```python
taux_attrition = df['Churn'][df['Churn'] == "Yes"].count()/df['Churn'].count()
taux_attrition
```

```python
df['gender'] = pd.factorize(df['gender'])[0]
df['Churn'] = pd.factorize(df['Churn'])[0]
df['Dependents'] = pd.factorize(df['Dependents'])[0]
df['PaperlessBilling'] = pd.factorize(df['PaperlessBilling'])[0]
df['Partner'] = pd.factorize(df['Partner'])[0]
df['PhoneService'] = pd.factorize(df['PhoneService'])[0]
```

```python
colonne = ['Contract', 'DeviceProtection', 'InternetService', 'MultipleLines', 'OnlineBackup', 'OnlineSecurity', 'PaymentMethod', 'StreamingMovies', 'StreamingTV', 'TechSupport']

for col in colonne :
  df_dummies = pd.get_dummies(df[colonne], prefix=colonne, dtype=float)

  # Je concat les deux DF pour n'en faire plus qu'un, puis je fais supprimer la colonne code_commune
  df = pd.concat([df, df_dummies], axis = 1)
  df.drop(colonne, axis = 1, inplace = True)


df.head()
```

```python
df.head()
```

```python
df.drop('customerID', axis = 1, inplace = True)
```

```python
ml_lr = LogisticRegression()

X = df.loc[:,df.columns != "Churn"]
y = df['Churn']

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size= 0.75, random_state=42)

ml_lr.fit(X_train,y_train)

ml_lr.score(X_test,y_test)
```
